## 2. PANDAS - DATA STRUCTURE

### 1. Series :-


```python
it is one dimentional array with index , it store a single column or row of data in dataframe

one dimentional array (labeled) capable of holding any type of data
```

### 2. DataFrame

```python
it is tabular spreadsheet like structure represneting rows each of which contain one or more columns

two dimentional data structured with columns of potentially different type of data
```

```python
df = pd.DataFrame({"name": ['Alfred', 'Batman', 'Catwoman'],
                   "toy": [np.nan, 'Batmobile', 'Bullwhip'],
                   "born": [pd.NaT, pd.Timestamp("1940-04-25"),
                            pd.NaT]})
```

```python
df.head()
```

```python
df.dropna()
```

```python
df.dropna(how='all')
```

```python
df.drop_duplicates(subset=['toy']) # remove dupkicate from toy
```

```python
df.drop_duplicates(subset=['toy'],keep='last') 
```

```python
df['name'].value_counts(dropna=False)
```
